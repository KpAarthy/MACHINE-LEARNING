```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ds=pd.read_csv('diabetes.csv')
ds
```

```python
ds1=ds.rename(columns={"BloodPressure":"BP","SkinThickness":"ST","DiabetesPedigreeFunction":"DPF",})
ds1
```

```python
ds1.shape
```

```python
ds1.head(10)
```

```python
sns.scatterplot(x=ds1['Pregnancies'], y=ds1['Glucose'], hue=ds1['Outcome'], style=ds1['BMI'])
plt.show()
```

```python
sns.pairplot(ds1, hue='Outcome',height=2)
plt.show()
```

```python
plt.hist(ds1['Pregnancies'], bins=20, rwidth=0.5, align='left', color='blue')
plt.show()
```

```python
sns.boxplot(ds1['Pregnancies'],orient='h', color='blue', order=7, linewidth=5)
plt.show()
```
